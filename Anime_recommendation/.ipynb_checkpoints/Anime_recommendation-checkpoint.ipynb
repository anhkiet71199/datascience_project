```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('./dataset/imdb_anime.csv')
data.head(5)
```

```python
print(data.shape)
print(data.info())
```

```python
# Change type column User Rating
data['User Rating'] = pd.to_numeric(data['User Rating'], errors='coerce')
```

```python
# Change type column Number of Votes
def replace_commas(x):
    try:
        return str(x).replace(',', '.')
    except:
        return x
data['Number of Votes'] = data['Number of Votes'].apply(replace_commas)
data['Number of Votes'] = pd.to_numeric(data['Number of Votes'], errors='coerce')
```

```python
data.isnull().sum()
```

## Data visualization

```python
movie_ratings_count = data.groupby('Title').size().sort_values(ascending=False)
plt.figure(figsize=(10,6))
movie_ratings_count[:20].plot(kind='bar')
plt.title("Top 20 movies by number of ratings")
plt.show()
```

```python
genres_split = data['Genre'].str.split(', ').explode()

# Plot the count of each genre
plt.figure(figsize=(10, 6))
sns.countplot(y=genres_split, order=genres_split.value_counts().index)
plt.xlabel('Count')
plt.ylabel('Genre')
plt.title('Count of Movies by Genre')
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
data.isnull().sum()
```

```python
data['Summary'].fillna('',inplace = True)
```

```python
# Preprocess and vectorize the text data
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(data['Summary'])
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
def get_anime_recommendations(title, cosine_sim=cosine_sim, top_n=5):
    idx = data.index[data['Title'] == title].tolist()[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    top_similar_anime = sim_scores[1:top_n+1]
    recommended_anime = [data.iloc[anime[0]]['Title'] for anime in top_similar_anime]
    return recommended_anime 

anime_title = 'Doraemon'
recommendations = get_anime_recommendations(anime_title)
print(f'Recommend for anime :  {anime_title}')
for i in recommendations:
    print(i)
```
