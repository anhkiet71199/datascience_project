```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('./dataset/Top_200_Movies_Dataset_2023(Cleaned).csv')
```

```python
print(data.head(5))
print(data.shape)
print(data.info())
```

```python
data['Release Date'] = pd.to_datetime(data['Release Date'])
```

```python
data['Total Gross'] = data['Total Gross'].str.replace('[\$,]','', regex=True)
data['Total Gross'] = pd.to_numeric(data['Total Gross'])
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

### Total Revenue by Distributor

```python
plt.figure(figsize=(10, 6))
sns.barplot(data = data,  x = 'Distributor', y = 'Total Gross', estimator='sum', errorbar=None)
plt.xticks(rotation=90)
plt.xlabel('Distributor')
plt.ylabel('Total Gross')
plt.title('Total Revenue by Distributor')
plt.show()
```

### Number of Movie by Distributor

```python
plt.figure(figsize=(10,8))
sns.countplot(data= data, x = 'Distributor')
plt.xticks(rotation=90)
plt.xlabel('Distributor')
plt.ylabel('Numner of Movie')
plt.title('Number of Movie by Distributor')
plt.show()
```

### Total Revenue by Release Month

```python
data['Release Month'] = data['Release Date'].dt.month
plt.figure(figsize=(10,8))
sns.barplot(data = data, x = 'Release Month', y = 'Total Gross', estimator='sum', errorbar = None)
plt.xticks(rotation=90)
plt.xlabel('Release Month')
plt.ylabel('Total Gross')
plt.title('Total Revenue by Release Month')
plt.show()
```

### Top 10 Highest-Grossing Movies

```python
data_sort = data.sort_values(by = 'Total Gross', ascending=False)

top_10_movies = data_sort.head(10)
bottom_10_movies = data_sort.tail(10)

plt.figure(figsize=(10,8))
plt.barh(top_10_movies['Title'], top_10_movies['Total Gross'])
plt.xlabel('Total Gross')
plt.ylabel('Movie')
plt.title('Top 10 Highest-Grossing Movies')
plt.gca().invert_yaxis()
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.barh(bottom_10_movies['Title'], bottom_10_movies['Total Gross'])
plt.xlabel('Total Gross')
plt.ylabel('Movie')
plt.title('Top 10 Lowest-Grossing Movies')
plt.show()
```
